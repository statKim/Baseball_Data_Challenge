# 1st Feature Engineering

```python
regular.iloc[4]
```

```python
regular.loc[(regular["batter_name"]=="강경학")&(regular["year"]==2016-1)]["OPS"].iloc[0]
```

```python
regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[4])&(regular["year"]==regular["year"][4]-1)]["OPS"].iloc[0]
```

```python
lag_1_OPS = []
for i in range(len(regular)):
    if len(regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"][i]-1)]["OPS"])==0:
        lag_1_OPS.append(np.nan)
    else:
        lag_1_OPS.append(regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"][i]-1)]["OPS"].iloc[0])
lag_1_OPS
```

```python
regular["lag_1_OPS"] = lag_1_OPS
```

```python
lag_1_avg = []
for i in range(len(regular)):
    if len(regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"][i]-1)]["avg"])==0:
        lag_1_avg.append(np.nan)
    else:
        lag_1_avg.append(regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"][i]-1)]["avg"].iloc[0])
regular["lag_1_avg"] = lag_1_avg
```

```python
regular.columns
```

```python
for j in ['avg', 'G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'TB', 'RBI', 'SB', 'CS', 'BB', 'HBP', 'SO', 'GDP',
         'SLG', 'OBP', 'E', 'OPS']:
    lag_1 = []
    for i in range(len(regular)):
        if len(regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"][i]-1)][j])==0:
            lag_1.append(np.nan)
        else:
            lag_1.append(regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"][i]-1)][j].iloc[0])
    
    regular["lag_1_" + j] = lag_1
    print(j)
```

```python
regular.columns
```

```python
# OPS 결측치 제거
regular = regular.loc[-regular["OPS"].isnull(),]
```

## Model building

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
y = regular["OPS"]
X = regular[['lag_1_avg', 'lag_1_G', 'lag_1_AB', 'lag_1_R',
       'lag_1_H', 'lag_1_2B', 'lag_1_3B', 'lag_1_HR', 'lag_1_TB', 'lag_1_RBI',
       'lag_1_SB', 'lag_1_CS', 'lag_1_BB', 'lag_1_HBP', 'lag_1_SO',
       'lag_1_GDP', 'lag_1_SLG', 'lag_1_OBP', 'lag_1_E', 'lag_1_OPS']]
```

```python
# tree 모델 500개
rf = RandomForestRegressor(n_estimators=500)
```

```python
# 결측값을 -1로 넣고 fit(결측치 있으면 error)
rf.fit(X.fillna(-1), y)
```

```python
train = regular.loc[regular["year"]<=2017,]
test =  regular.loc[regular["year"]==2018,]
y_train = train["OPS"]
X_train = train[['lag_1_avg', 'lag_1_G', 'lag_1_AB', 'lag_1_R',
       'lag_1_H', 'lag_1_2B', 'lag_1_3B', 'lag_1_HR', 'lag_1_TB', 'lag_1_RBI',
       'lag_1_SB', 'lag_1_CS', 'lag_1_BB', 'lag_1_HBP', 'lag_1_SO',
       'lag_1_GDP', 'lag_1_SLG', 'lag_1_OBP', 'lag_1_E', 'lag_1_OPS']]

y_test = test["OPS"]
X_test = test[['lag_1_avg', 'lag_1_G', 'lag_1_AB', 'lag_1_R',
       'lag_1_H', 'lag_1_2B', 'lag_1_3B', 'lag_1_HR', 'lag_1_TB', 'lag_1_RBI',
       'lag_1_SB', 'lag_1_CS', 'lag_1_BB', 'lag_1_HBP', 'lag_1_SO',
       'lag_1_GDP', 'lag_1_SLG', 'lag_1_OBP', 'lag_1_E', 'lag_1_OPS']]
```

```python
# 2017 이전 데이터로 train
rf = RandomForestRegressor(n_estimators=500)
rf.fit(X_train.fillna(-1), y_train, sample_weight=train["AB"])
```

```python
pred = rf.predict(X_test.fillna(-1))
```

```python
real = test["OPS"]
ab = test["AB"]
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(real, pred, sample_weight=ab)**0.5
```
